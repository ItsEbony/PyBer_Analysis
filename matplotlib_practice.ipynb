```python
%matplotlib inline
import matplotlib.pyplot as plt
import statistics
import numpy as np
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)

plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot. 
plt.plot(x_axis, y_axis, color="blue")
stdev = statistics.stdev(y_axis)
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
stdev
plt.show()
```

```python
# Create the plot and add a lable for the legend. 
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes 
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Set the y limit between 0 and 45. 
plt.ylim(0, 45)
# Create a title 
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend. 
plt.legend()
```

```python
# Create the plot and add a lable for the legend. 
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label='Boston')
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3, color="green")
# Create labels for the x and y axes 
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
#Set the y limit between 0 and 45. 
ax.set_ylim(0, 45)
# Create a title 
ax.set_title("PyBer Fare by Month")
# Add a grid
ax.grid()
# Add the legend. 
ax.legend()
```

```python
#Create and annotate bar chart
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3, color="green", label='Boston')
#Create labels for x and y
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Create a title
plt.title("PyBer Fare by Month")
#Add the legend
plt.legend()
```

```python
#Create and annotate bar chart
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
#Create labels for x and y
plt.xlabel("Fare($)")
plt.ylabel("Date")
#Create a title
plt.title("PyBer Fare by Month")
#Add the legend
plt.legend()
```

```python
# Create plot using object-oriented approach
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Create plot using object-oriented approach (horizontally this time!!)
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.set_xticks(np.arange(0, 51, step=5.0))
ax.invert_yaxis()
#Labels/Title
ax.set_xlabel("Fares ($)")
ax.set_ylabel("Date")
ax.set_title("Pyber Fare by Month")
#Add legend
ax.legend()
```

```python
plt.subplots(figsize=(8, 8)) #order matters here, this has to go above pie chart
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
fig, ax = plt.subplots(figsize=(8, 8)) #okay, so to change size put in this para
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,labels=x_axis, colors=colors, startangle=90, shadow=True, explode=explode_values, autopct='%.1f%%')
ax.invert_xaxis()
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
